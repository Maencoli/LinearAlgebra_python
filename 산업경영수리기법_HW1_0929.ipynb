# 문제 2.7

Write a program which simulates the followings. 
Pick an initial point z0 = (x0, y0).t arbitrarily in the plane. For k = 0, 1, 2, ..., set

z_{k+1} := (x_{k+1}, y_{k+1}) = A_{sk}z_{k} + d_{sk}
where sk is chosen randomly from the set {1, 2, ..., m}, with  probability p_{i} = Pr(sk = i), i = 1, ..., m and \sigmaP_{i} = 1

A is a (2,2)-matrix and d is a 2-D column vector.). 
Plot the sequence. 

## (a) What can you see in the figure generated by your program for the following data set?

```python
##data set:###
import numpy as np

A = {
    1: np.matrix('0 0; 0 .27'),
    2: np.matrix('-.139 .263; .246 .224'),
    3: np.matrix('.17 -.215; .222 .176'),
    4: np.matrix('.781 .034; -.032 .739')
     }
d = {
    1: np.matrix('.5; 0'),
    2: np.matrix('.57; -.036'),
    3: np.matrix('.408; .0893'),
    4: np.matrix('.1075; .27')
}
p = {
    1: .02,
    2: .15,
    3: .13,
    4: .70
}
```

```python
import random

def matCalc(z, n, k): #n = 몇개 뽑을지, k = set
    ###1. choose S_{k} with probability p
    k_sample = list(range(k))
    S_k = random.choices(k_sample, weights=(p[1],p[2],p[3],p[4] ), k=1)[0]
    plot = []

    while k < n:
        ###2. Calculate A_{S_k} z_{k} + d+{S_k}
        z_next = A[S_k] @ z + d[S_k]
        print(z)
        #print(plot)
        #plot.append(k)
        k += 1
        return matCalc(z_next, n, k) 
    return plot
```

```python
z_0 = np.matrix('0;0')
matCalc(z_0, 30, 4)
```
